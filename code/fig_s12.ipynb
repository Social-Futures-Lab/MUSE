```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib

path = '../data/'

plt.style.use('default')
plt.rcParams['font.family'] = 'Helvetica'
matplotlib.rcParams['font.size'] = 12
font_size_small = 10.5
```

```python
data_all = pd.read_csv(path + "notes_all.csv", dtype=str)
scores_all = [float(s) for s in data_all["note_score"] ]
mean_score_all = np.mean(np.array(scores_all))
std_score_all = np.std(np.array(scores_all))

data_eval_um = pd.read_csv(path + "responses_unimodal.csv", dtype=str)
data_eval_mm = pd.read_csv(path + "responses_multimodal.csv", dtype=str)

scores_high = [float(s) for s in data_eval_um['note_1_score']] + [float(s) for s in data_eval_mm['note_1_score']]
mean_score_high = np.mean(scores_high)
std_score_high = np.std(scores_high)

scores_avg = [float(s) for s in data_eval_um['note_2_score']] + [float(s) for s in data_eval_mm['note_2_score']]
mean_score_avg = np.mean(np.array(scores_avg))
std_score_avg = np.std(np.array(scores_avg))
```

```python
fig, ax = plt.subplots(figsize=(7.5, 2.8))
plt.hist(scores_all, label="All", color='#d7d7d5')
plt.hist(scores_avg, label="Average helpfulness", color='#F7E6CC', bins=4)
plt.hist(scores_high, label="High helpfulness", color='#F8D3A6', bins=4)
plt.xlabel("Helpfulness scores")
plt.ylabel("Number of responses")
plt.yticks(fontsize=font_size_small)
plt.legend(loc='upper right', facecolor='w', edgecolor='w', framealpha=1, fancybox=False,
           handlelength=0.7, handletextpad=0.5, borderpad=0.2, labelspacing=0.25, columnspacing=1,
           bbox_to_anchor=(0.855, 1.2), ncol=3)
plt.axvline(x=mean_score_all, color='k', linestyle='--') 
plt.axvline(x=mean_score_avg, color='#e28413', linestyle='--') 
plt.axvline(x=mean_score_high, color='#ca5310', linestyle='--') 
plt.text(0.01, 1000, str(round(mean_score_all, 2))+r'$\pm$'+str(round(std_score_all, 2)), color='k')
plt.text(0.18, 95, str(round(mean_score_avg, 2))+r'$\pm$'+str(round(std_score_avg, 2)), color='#e28413')
plt.text(0.45, 95, str(round(mean_score_high, 2))+r'$\pm$'+str(round(std_score_high, 2)), color='#ca5310')
plt.setp(ax.spines.values(), color='k', linewidth=0.75)
plt.yscale('log')
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.yaxis.set_tick_params(width=0.75)
```
